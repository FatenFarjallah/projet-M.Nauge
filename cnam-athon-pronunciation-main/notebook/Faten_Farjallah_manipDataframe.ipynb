```python
#importer le fichier Buchanan
dfDicoBuch = pd.read_csv("./../data/dictionary/1757_Buchanan-J.csv", delimiter=";", encoding="utf8")
dfDicoBuch.sample(n=5)
```

```python
#importer le fichier PronChar
PronChar = pd.read_csv("./../data/dictionary/Buchanan_PronChar_counts.csv", delimiter=";", encoding="utf8")
PronChar.sample(n=5)
```

```python
PronChar.fillna('', inplace=True)
```

```python
PronChar.sample(n=5)
```

```python
def char2graph(strChar, dfMapping):
    res = ""
    
    for valChar in strChar:
        #print(valSampa)
        dfRowFinded = dfMapping[dfMapping['char']==valChar]

        if len(dfRowFinded==1):
            res+=dfRowFinded.iloc[0]['graph']
        else:
            res+='?'
    return res
```

```python
dfDicoBuch['motVedette'] = dfDicoBuch.apply(lambda x: char2graph(x['pron'], PronChar),1)
```

```python
#dfDicoBuch['motVedette']
dfDicoBuch.sample(n=10)
```

```python
#### 2.1.Résolution du probléme :
```

```python
#puisqu'on a que 1000 mots dans le ficher BigiRichSample, on va utiliser le dictionnaire à sa place pour remplir "Sampa" et "ipa"
dfDicoBigi = pd.read_csv("./../data/dictionary/2015_Bigi-B.dict", delimiter=" \[\] ", encoding="utf8", engine='python', names=['hwd','pronSampa'], skiprows=8 )
dfDicoBigi.sample(5)
```

```python
#jointure
tableF=pd.merge(dfDicoBuch, dfDicoBigi,left_on='motVedette', right_on='hwd',how='left')
tableF.sample(5)
```

```python
dfSampaIpa = pd.read_csv("./../data/dictionary/sampa-ipa.txt", delimiter="  *", encoding="utf8", engine='python', names=['sampa','ipa'])
dfSampaIpa.drop_duplicates(subset=['sampa'], keep='first', inplace=True)
dfSampaIpa.sample(n=5)
```

```python
tableF.fillna('', inplace=True)
```

```python
def sampa2ipa(strSampa, dfMapping):
    ipaRes = ""
    
    listValSampa = strSampa.split(" ")
    #print(listValSampa)
    for valSampa in listValSampa:
        #print(valSampa)
        dfRowFinded = dfMapping[dfMapping['sampa']==valSampa]

        if len(dfRowFinded==1):
            ipaRes+=dfRowFinded.iloc[0]['ipa']
        else:
            ipaRes+='?'
    return ipaRes
```

```python
tableF['ipa'] = tableF.apply(lambda x: sampa2ipa(x['pronSampa'], dfSampaIpa),1)
```
