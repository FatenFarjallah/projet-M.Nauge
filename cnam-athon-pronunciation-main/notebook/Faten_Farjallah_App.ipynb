```python
# import necessaire à la création des widgets
from IPython.display import display, clear_output, HTML, Markdown 
import ipywidgets as widgets
import re
from IPython import display as dis
###
import pandas as pd
##
import requests_html
from requests_html import HTMLSession
import requests
session = HTMLSession()
from bs4 import BeautifulSoup
```

```python
#importer le fichier 
BuchananRich = pd.read_csv("./../data/dataframe/BuchananRich.csv", delimiter=";", encoding="utf8")
BuchananRich.sample(5)
```

```python
#supprimer les NAN
BuchananRich=BuchananRich.fillna("")
BuchananRich.sample(5)
```

```python
#creer les filtres  
filtre_hwd = widgets.Text(value="",description='filtre hwd',layout={'width': '300px'} )
filtre_ipa = widgets.Text(value="",description='filtre ipa',layout={'width': '300px'} )
filtre_occ = widgets.IntSlider(name='occ.Max', min=1, max=20, step=1, value=3)
```

```python
def filtDf(strHwd,strIpa,intOcc):    
    df = BuchananRich[BuchananRich['motVedette'].str.contains(strHwd,na=False,flags=re.IGNORECASE, regex=True)] 
    df=df[df['ipa'].str.contains(strIpa, na=False,flags=re.IGNORECASE, regex=True)]
    df = df.head(intOcc)
    return df
```

```python
# définition du widget button permettant de lancer l'extraction
button = widgets.Button(description='Collecter')

# définition du widget permettant l'affichage d'une sortie
out1 = widgets.Output()
out2 = widgets.Output()

# spécifier les actions à réaliser lors de l'appui sur le bouton
def on_button_clicked(_):
    
    # récupérer le texte saisi par l'utilisateur
    strHwd = filtre_hwd.value
    strIpa=filtre_ipa.value
    intOcc = int(filtre_occ.value)
    
    print(strIpa)

    with out1:
        clear_output()
        df= filtDf(strHwd, strIpa,intOcc)
        tableau ='<table> \
                         <tr>\
                             <th>motVedette</th>\
                             <th>nbSyll</th>\
                             <th>pron</th>\
                             <th>pronSampa</th>\
                             <th>ipa</th>\
                        </tr>'
        
    
        for i in range(len(df)):
    
            tableau += '<tr>\
                             <td>'+df.motVedette.iloc[i]+'</td>\
                             <td>'+str(df.nbSyll.iloc[i])+'</td>\
                             <td>'+df.pron.iloc[i]+'</td>\
                             <td>'+df.pronSampa.iloc[i]+'</td>\
                             <td>'+df.ipa.iloc[i]+'</td>\
                        </tr>'
                        
        tableau += '</table>'
        
   
        display(HTML("<br/>"+tableau))


    with out2:
        clear_output()
        for i in range(len(df)):
            
            targetWord = df.motVedette.iloc[i]
            urlLongman = "https://www.ldoceonline.com/dictionary/"
            urlTarget = urlLongman+targetWord+'.html'
            #recuperer les url
            response = session.get(urlTarget)
            soup = BeautifulSoup(response.text,'html.parser')
            souplettesBody = soup.find('body')
            #url British
            findedB = souplettesBody.find('span',{"class":"speaker brefile fas fa-volume-up hideOnAmp"})
            lienB=findedB['data-src-mp3']
            #url American
            findedA = souplettesBody.find('span',{"class":"speaker amefile fas fa-volume-up hideOnAmp"})
            lienA=findedA['data-src-mp3']
            #suprimer la version
            lienB= lienB.replace(lienB[-15:],"")
            lienA = lienA.replace(lienA[-15:],"")

            
            
            #afficher les audios 
            audio_widgets = []

            listUrl = []

            listUrl.append(lienB)
            listUrl.append(lienA)
            
            i=0
            display(HTML(targetWord + "<br/>"))
            for urlCur in listUrl:
                if i==0:
                    display(HTML("Pron. UK :" ))
                    i=1
                else:
                    display(HTML("Pron. US :" ))
                    i=0
                display(dis.Audio(url=urlCur))
            display(HTML("<br/>" ))
                

                
button.on_click(on_button_clicked)
```

```python
# les widgets de filtre input
vbFiltres = widgets.HBox([filtre_hwd,filtre_ipa,filtre_occ,button])

# les widgets d'output
vbOuts = widgets.VBox([out1,out2])

hbAllWidgets = widgets.VBox([vbFiltres,vbOuts])
hbAllWidgets
```
